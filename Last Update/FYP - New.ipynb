```python
import yfinance as yf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import pickle
from keras.models import load_model
import joblib
```

```python
lst = ['ADANIPORTS', 'ASIANPAINT', 'AXISBANK', 'BAJAJ-AUTO', 'BAJAJFINSV',
       'BAJFINANCE', 'BHARTIARTL', 'BPCL', 'BRITANNIA', 'CIPLA',
       'COALINDIA', 'DRREDDY', 'EICHERMOT', 'GAIL', 'GRASIM', 'HCLTECH',
       'HDFC', 'HDFCBANK', 'HEROMOTOCO', 'HINDALCO', 'HINDUNILVR',
       'ICICIBANK', 'INDUSINDBK', 'INFY', 'IOC', 'ITC', 'JSWSTEEL',
       'KOTAKBANK', 'LT', 'M&M', 'MARUTI', 'NESTLEIND', 'NTPC', 'ONGC',
       'POWERGRID', 'RELIANCE', 'SBIN', 'SHREECEM', 'SUNPHARMA',
       'TATAMOTORS', 'TATASTEEL', 'TCS', 'TECHM', 'TITAN', 'ULTRACEMCO',
       'UPL', 'VEDL', 'WIPRO', 'ZEEL']
```

```python
scaler =  MinMaxScaler(feature_range = (0,1))
for x in lst:
    
    #Preprocessing
    data = pd.read_csv('./Data/' + x + '.csv')
    df = list(data[['Open','High','Low', 'Close']])
    df_train = data[df]
    df_for_training_scaled = scaler.fit_transform(df_train)
    trainX = []
    trainY = []
    n_future = 1
    n_past = 14
    for i in range(n_past, len(df_for_training_scaled) - n_future +1):
        trainX.append(df_for_training_scaled[i - n_past:i, 0:df_train.shape[1]])
        trainY.append(df_for_training_scaled[i + n_future - 1,:])
    trainX, trainY = np.array(trainX), np.array(trainY)
    print(trainY.shape)
    print(trainX.shape)
    
    # LSTM Model
    model = Sequential()
    model.add(LSTM(64, input_shape = (trainX.shape[1], trainX.shape[2]), return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(32, return_sequences = False))
    model.add(Dense(4))
    model.compile(optimizer = 'adam', loss = 'mse')
    model.summary()
    model.fit(trainX, trainY, epochs = 200, batch_size = 32, validation_split = 0.15, verbose = 1)

    model.save('./Models1/' + x + '.h5')
    model = load_model('./Models1/' + x + '.h5')
    tmp1 = []
    
    # Prediction
    for i in range(365):
        pred = model.predict(trainX[-1:])
        tmp = np.append(trainX[-1,-13:], pred, axis = 0)
        trainX = np.append(trainX, [tmp], axis = 0)
        tmp1.append(scaler.inverse_transform(pred)[0])
        
    # Plotting
    plt.figure(figsize = (20, 10))
    plt.plot(tmp1)
    plt.xlabel('Days')
    plt.ylabel('Stock Price')
    plt.title(x)
    print(x)
    plt.savefig('./Plots/' + x + '.png')
    plt.show()
```
